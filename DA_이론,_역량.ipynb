### 데이터 분석 방법론 : CRISP-DM (CRoss Industry Standard Process for Data Mining)

**1) Business Understanding**
- 비즈니스 배경 및 목적 (비즈니스가 처한 상황, 문제점, 사업목표, 조직구성 등)   
`ex) 대리점 매출저하로 인한 사업성장문제를 개선하기 위한 방법`
- 분석 목적 및 목표 설정 (데이터분석 이유, 목적, 효과)   
`ex) 비즈니스 목적이 매출을 증대시키는 것 이라면 분석 목적은 증대 방법을 탐색하는 모델 수립`


**2) Data Understanding**
- 초기데이터 수집 : 고객데이터, 상품데이터, 구매데이터, 로그데이터 등
- 데이터 기술(describe) : 확보데이터에 대한 설명 자료   
`ex) 테이블 별 데이터 수, 테이블 항목 데이터유형(categoric, iterator, integer, numeric, character)`

**3) Data Preprocessing**
- 데이터누락, 오류, 의미해석불가 데이터들 =>정제
- 정제된 데이터로 새로운 데이터를 생성하거나 통합
- 데이터에 대한 이해를 통해 데이터 종류, 건수, 분석기간, 분석기준을 확정   
> **Tip** : 활용가능한 모든 데이터를 분석하기보다 목적에 맞게 최적화하여 분석(F검정 후 t검정을 통과하는 정도면 충분)

**4) Modeling**
- 모델 선정 가이드
   - 데이터량이 충분한가?(신뢰도)
   - 데이터의 품질이 중요한 모델인가?
   - 데이터의 형태가 모델에 적합한가? (변환을 하면 되지만, 성능에 영향을 줄 수 있다)
- 테스트 방안 설계,수립
   - 모델링 평가지표와 basline을 정해두어야한다.  
ex) 지도학습의 오류율 / 비지도학습의 오차값평균, 중심점, 실루엣   
    

**5) Evaluation**
- 모델의 성능이 목표와 일치 하는지, 비즈니스에 어떤 의미와 효과가 있을지 정리
- 평가지표 분석 및 모델성능 최적화(반복)


**6) Deployment(배포)**
- 결과 발표 및 보고 단계
- 데이터와 비즈니스에 대한 이해, 전처리단계, 모델링기술, 결과와 시사점을 정리 
- 향후 자동화 구축방안과 모니터링 방안을 추가

## 문제를 쪼개어 분석 : MECE(Mutually Exclusive Collectively Exhaustive)

**문제해결을 위한 이슈 분석**

**1. 핵심질문을 정의한다**
   - 과거,현재,미래 상황(배경)과 현재 직면한 문제 및 장애요소 정의

**2. 핵심질문을 이슈로 바꾼다.**
   - 핵심질문을 이슈로 발전시켜서 실천적이고 의사결정에 도움이 되도록 구체화

**3. 이슈를 잘게 쪼갠다.(MECE)**
   - 이슈를 상위이슈와 하위이슈(인과관계나 해결책)의 관계로 구성된다.
   - 하위 이슈 상호간에는 중복과 누락이 없어야한다.   
   `ex) 사람을 남자, 여자로 나누면 중복이없고, 누락되는 사람이 없다`
   - 하위 이슈의 단계는 충분히 검증 가능한 데이터 수집이나 분석을 할 수 있는 수준까지 나눈다.

**4. 쪼갠 이슈를 해결할 가설을 세우고 근거를 파악한다.**
- 도출된 이슈에 대한 가설을 미리 상정하고 이를 입증하기 위한 분석을 진행해야한다.
**5. 작업계획을 수립한다.**

## 가설수립 - 연역적 추론과 귀납적 추론

### 연역적 추론(Inductive Reasoning)


- 미리 정해놓은 예상 결과를 검증하는 방법   
- 데이터와 산출된 정보는 이론을 증명하고 검증하는 **수동적 역할**
- 가설의 **예외적인 결과에 대해서 설명하기 어렵다**는 단점이 있다.   
   `ex) 가설 : 매일 댓글을 다는 사용자는 액티브유저이다`   
     `-> 매일 접속은 하지만 댓글만 안다는 유저는 설명불가`

- **연구단계** : 배경지식 -> 가설 -> 연구설계 -> 측정 -> 자료수집 -> 분석 -> 경험적 일반화
   - 직접추론 : P이면 Q이다 -> Q이면 P이다(대우명제)
   - 간접추론 : 모든 사람은 죽는다 -> 나는 사람이다 - > 나는 죽는다.   



### 귀납적추론(Deductive Reasoning)

- 어떤 현상을 관찰하여 특성을 모두 열거하고 측정하여 유의미한 결과를 일반화 하는 방법(패턴찾기)
- 데이터는 이론을 새롭게 발견하고 명료화 하는 **능동적 역할**
- 데이터 분석가의 직관이 개입될 여지가 적다.
- **인과관계의 오류가 발생할 수 있고 일반화가 어렵다.**   
`ex) 맹자도 죽었다, 공자도 죽었다. 그러므로 사람은 죽는다`   
`ex) 비둘기는 난다, 참새도 난다, 그러므로 모든 새는 난다 => 일반화 모순`
- **연구단계** : 경험적 관찰 -> 가설 -> 연구설계 -> 측정 -> 자료수집 -> 분석 -> 이론수립

> **연역적 추론이 귀납적 추론보다 효율성이 높다.**
- 경험주의적 사고를 바탕으로하기 때문에 가설 생성에 소요되는 시간이 적다.
- 하지만 최근에는 기술의 발달로 데이터 수집과 처리가 빨라지면서 많은 데이터를 필요로하는 귀납적추론의 단점이 해소되는 상황

## 퍼널분석(Funnel Analysis)

- 설계자가 설계한 **유저경험루트를 고객이 잘 따라오는지** 최초유입부터 단계까지 나누어 분석하는 기법
- 분석을 각 단계별 **이탈고객 수**를 확인 할 수 있다.
- 각 단계를 통과할 때 마다 유저수가 줄어들게 되어 깔데기(Funnel)분석이라한다.
- 각 단계를 넘어가는 것을 **전환(Conversion)**이라고 부르고 그 비율을 **전환율(Conversion Rate)**

### 이항분포 (Binomial Dstribution)

- 독립적으로 반복되는 행위에 의해서 두가지 결과를 가지는 사건 ex) 동전 던지기

- k개의 앞면이나오고, n-k개의 뒷면이 나올 확률 $p^k(1-p)^{(n-k)}$   

$$n! / k!(n-k)! \times {p^k}{(1-p)^{(n-k)}}$$

### 조건부 확률(Conditional Distribution)

- 어떤 사건의 결과에 영향을 받는 한 사건의 결과에 대한 확률(B가 일어나면서 A가 일어날 확률)

$$P(A|B) = P(A \cap  B) / P(B) = P(B|A)P(A) / P(B)$$
- P(B) = P(B|A) x P(A) +  P(B|^A) x P(^A)
- P(A⋂B) = P(A|B) x P(B)
- P(B|A) = 가능도 / P(A) = 사전확률 / P(B) = 증거


ex) 전체 인구 10%가 걸리는 암, 암이 있는 사람중 90%는 positive, 없는 사람중 80% 는 negative   
- P(C) = 0.1
- P(pos | C) = 0.9
- P(neg |^C) = 0.8


유추할 수 있는 값   
- P(^C) = 0.9
- P(neg | C) = 0.1
- P(pos |^C) = 0.2

### 모집단

- 모집단(Population) : 다루려는 대상의 모든 데이터
- 모수 : 모집단의 데이터를 설명하는 수치값(모평균, 모분산, 모표준편차, 모비율, 모상관관계


- 표본(sample) : 모집단에서 추출된 부분집합
- 통계량(Statsitics) : 표본평균, 표본분산, 표본표준편차, 표본비율, 표본상관관계
- 표본분포(Distribution of a Statsitics) : 표본집단의 통계량 확률 분포

- 기술통계 : 우리가 수집한 데이터를 요약,묘사, 설명하는 통계기법
- 추리통계 : 수집한 데이터를 바탕으로 모집단에 대해 추론하는 통계기법

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

np.random.seed(10)

# 1과 0중에서 5000개를 22%와 78%의 확률로 추출
a = np.random.choice([0,1], size=5000, p=[.22, .78])

# 모집단
print(f'비율 : {Counter(a)}')
print(f'평균 : {a.mean()}')
print(f'분산 : {a.var()}')
print(f'표준편차 : {a.std()}')

# 모집단의 분산 = 모집단의 평균 x (1-모집단의 평균)
# a.mean() * (1 - a.mean())
```

### 부트스랩(Bootstrapping)

- 표본을 모집단으로 가정한 뒤, for loop을 사용하여 샘플들의 평균 분포를 확인
- 복원추출로 원하는 개수의 데이터를 추출(ex - 10개의 데이터를 10000번 추출)
- 표본평균의 크기는 모집단평균의 크기와 비슷하다
- 표본분산의 크기는 (모집단의평균 x (1-모집단의평균))/샘플 사이즈
- 샘플의 사이즈(10개 ->100개)가 클수록 표본평균의 분포가 모집단의 평균에 가까워진다.(= 분산이 줄어든다)

```python
# a목록의 10개를 10000번 복원 추출하여 평균을 sample list에 넣기
import seaborn as sns
sample_10 = []

for _ in range(10000):
    sample = np.random.choice(a, 10, replace=True)
    sample_10.append(sample.mean())

plt.hist(sample_10, alpha = 0.5, label= 'sample_10')
# sns.kdeplot(sample_10, alpha = 0.5, label= 'sample_10')
# sns.displot(sample_10, alpha = 0.5, label= 'sample_10')
plt.show();
```

```python
# a목록의 100개를 10000번 복원 추출하여 평균을 sample list에 넣기
sample_100 = []

for _ in range(10000):
    sample = np.random.choice(a, 100, replace=True)
    sample_100.append(sample.mean())
    
# print(sample_10)
# print(sample_100)

# 모수의 평균, 샘플 10개와 100개의 분포 
plt.hist(sample_10, alpha = 0.5, label= 'sample_10')
plt.hist(sample_100, alpha = 0.5, label= 'sample_100')
plt.axvline(x=a.mean(), c = 'red', label= 'Population Mean' )
plt.legend()
plt.show();
```

**큰수의 법칙(Law of Large Numbers)**
- 샘플사이즈(표본의 크기)가 커질수록 샘플의 통계치는 모집단의 모수에 가까워진다.

**중심 극한 정리 (Central Limit Theorem)**
- 모집단의 분포에 상관없이 임의의 분포에서 추출된 표본들의 평균 분포는 정규분포를 이룬다.
- 표본을 추출할 때 충분한 샘플사이즈를 추출해야한다.
- 모집단이 skewd(꼬리)되어 있어도 적절한 샘플사이즈를 가지면 표본평균 분포는 정규분포를 띈다.

```python
sample_10 = np.array(sample_10)

# 표본 평균은 모집단의 평균과 비슷하고, 
# 표본 분산은 모집단의 분산을 샘플사이즈(10)로 나눈 값이다.
print(f'표본의 평균 : {sample_10.mean()}')
print(f'표본의 분산 : {sample_10.var()}')
print(f'표본의 표준편차 : {sample_10.std()}')

# 표본 분포의 분산은 (모집단의 평균 * (1-모집단의평균))/ 샘플 사이즈
```

**신뢰 구간 (Confidence Interval)**
- 모수에 대한 특정값을 예측하지 않고 모수를 포함하고 있는 구간을 제공 하는데, 이 구간이 신뢰구간이다.
- 불확실성을 해소할 수 있다.
- 모수가 신뢰구간안에 있을 확률이다.(표본의 분포와는 상관없음)

```python
sample_100 = np.array(sample_100)
sample_100_mean = sample_100.mean()
sample_100_mean
```

```python
lower, upper = np.percentile(sample_100, 2.5), np.percentile(sample_100, 97.5)
lower, upper
```

```python
plt.hist(sample_100, alpha=0.7)
plt.axvline(lower, c='red',label='95% Confience Interveal')
plt.axvline(upper, c='red')
plt.axvline(sample_100_mean, c='green',label='sample_100_mean')
plt.legend()
plt.show();
```

```python
# 신뢰구간을 plt.bar로 표현해보기
from scipy.stats import t
from scipy import stats

# np.random.seed(2)

sample = []
for _ in range(100):
    s1 = np.random.choice([0.1, 0.005, 0.02, 0.04, 0.214, 0.024, 0.075],100, replace=True)
    sample.append(s1.mean())

sample = np.array(sample)   # 표본
sample_mean = sample.mean() # 표본평균
n = len(sample)                 # 표본 개수
stderr = stats.sem(sample)      # 표준오차

confidence = 0.95           # 신뢰구간

# 신뢰구간 구하기
# stats.t.ppf((1+신뢰구간)/2, 표본개수-1, loc=평균, scale=표준편차)

# 평균에서 떨어진 거리 (bar에 yerr= 할때)
interval = stderr * stats.t.ppf( (1+confidence)/2, n-1)

# hist에 axvline 표시할 때
lower = sample_mean - interval # 오차 하한
upper = sample_mean + interval # 오차 상한
```

```python
plt.hist(sample, alpha=0.7)
plt.axvline(lower, c='red',label='95% Confience Interveal')
plt.axvline(upper, c='red')
plt.axvline(sample_mean, c='green',label='sample_mean')
plt.legend()
plt.show();
```

```python
plt.bar(x='sample', height=sample_mean, yerr=interval, capsize=10) # yerr = 오자범위, capsize= 선의 길이
plt.axhline(sample_mean, color = 'black', xmin = 0.4, xmax = 0.6) # 가운데 선의 길이
plt.show()
```
