### **2-1. Norm**

- ì£¼ì–´ì§„ ê°’ $x_0, x_1, ... x_n$ ë¥¼ elementë¡œ ê°–ëŠ” $\vec v$ì— ëŒ€í•˜ì—¬ 
  - $L_1$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤. 
    > $|\vec v| = |x_0| + |x_1| + ... + |x_n|$ <br>
  - $L_2$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
    > $||\vec v|| = \sqrt {{x_0}^2 + {x_1}^2 + ... + {x_n}^2}$
- ì´ë¥¼ ì°¸ì¡°í•˜ì—¬ ì£¼ì–´ì§„ ë²¡í„°ì— ëŒ€í•´ì„œ $L_2$ Norm ì˜ ê°’ê³¼ $L_1$ Norm ì˜ ê°’ì˜ ì°¨ì˜ ì ˆëŒ€ê°’ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“œì„¸ìš”.

##### **Q3. $\vec v = [3, -1, 8, 9]$ì¼ ë•Œ, NormDif($\vec v$)ì˜ ê²°ê³¼ê°’ì„ ë°˜ì˜¬ë¦¼í•˜ì—¬ ì†Œìˆ˜ì  ë‘˜ì§¸ ìë¦¬ê¹Œì§€ ì“°ì„¸ìš”.**

```python
def NormDif(v):
 L1 = sum([abs(x) for x in v])
 L2 = np.sqrt(sum([(x**2) for x in v]))
 return np.abs(L2 - L1)
round(NormDif([3,-1,8,9]),2)
```

##### **Q4. ì˜ˆì¸¡ê°’ $x = [2085, 1815, 2235, 1400, 2500, 1430, 3070, 2000, 1299, 1180]$ ì´ê³ , ì‹¤ì œê°’ $y=[2103, 2099, 1985, 1630, 2480, 1300, 3200, 1990, 1405, 1210]$ ì¼ ë•Œ, error í•¨ìˆ˜ë¥¼ í†µí•´ ì–»ì€ MSEì™€ MAE ê²°ê³¼ê°’ì„ (MSE ê°’, MAE ê°’) í˜•íƒœë¡œ ì“°ì„¸ìš”.**

```python
from typing import MutableSequence
def error(x, y, type):
  if type == 'MSE':
    value = ((x-y)**2).mean()
  elif type == 'MAE':
    value = abs(x-y).mean()
  else:
    print("ì—†ìŒ")
  return value
x=np.array([2085,1815,2235,1400,2500,1430,3070,2000,1299,1180])  
y=np.array([2103,2099,1985,1630,2480,1300,3200,1990,1405,1210])
print(F"MSE :{error(x,y,'MSE')},MAE:{error(x,y,'MAE')}")
```
